# Supervised vs Unsupervised Learning

Today I explored the two major categories of Machine Learning.

Goal:
- Understand Supervised Learning
- Understand Unsupervised Learning
- Implement one example of each

MAJOR DIFFERENCE I EXPLORED:

SUPERVISED LEARNING:It used features and targets to predict the model.

UNSUPERVISED LEARNING:It is used to find patterns,make groups and it includes no targets.

## Example 1: Student Score Prediction

Problem:
Can we predict a student's score based on hours studied?

```python
import pandas as pd
STUDENT=pd.DataFrame({"Hours":[2,4,6,8],"Marks":[30,50,70,90]})

STUDENT
```

Hours = Feature

Marks = Target

Reason:
Hours Studied helps predict Marks.

This is Supervised Learning because the answer column exists.

```python
import matplotlib.pyplot as plt

plt.scatter(STUDENT["Hours"],STUDENT["Marks"])
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

 **# Train-Test Split

Purpose:
To evaluate how well a model performs on unseen data.**

```python
from sklearn.model_selection import train_test_split
x=STUDENT[["Hours"]]
y=STUDENT["Marks"]
x_trian,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_trian.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**REFLECTION**

Why Train-Test Split?

A model should not be tested on the same data it learned from.

Training Data:
Used to learn patterns.

Testing Data:
Used to evaluate performance on unseen data.

```python
CUSTOMER=pd.DataFrame({"Income":[10,20,30,40,50],"Spending":[15,25,35,45,55]})

CUSTOMER
```

No target column exists.

This is Unsupervised Learning because the model must discover patterns itself.
The goal is not to predict,but to make relative groups and finding patterns.


| Supervised Learning | Unsupervised Learning |
|---------------------|-----------------------|
| Has Target Variable | No Target Variable |
| Learns from Answers | Finds Hidden Patterns |
| Predicts Outcomes | Groups Similar Data |
| Example: Marks Prediction | Example: Customer Segmentation |

What I Learned Today;

1. Machine Learning is divided into Supervised and Unsupervised Learning.

2. Features are the input variables.

3. Target is the output variable we want to predict.

4. Train-Test Split helps evaluate a model on unseen data.

5. Supervised Learning predicts outcomes.

6. Unsupervised Learning discovers patterns.
